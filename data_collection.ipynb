## Dataset 2

```python
df2 = pd.read_csv("./data/DiseaseAndSymptoms.csv")
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2["Disease"].value_counts()
```

```python
# Disease and symptoms

from collections import defaultdict

diseases = defaultdict(list)
features = df2.columns

for idx, row in df2.iterrows():

    disease = row["Disease"]

    symptoms = []

    for feature in features:

        if feature.startswith("Symptom") and str(row[feature]) != "nan":
            symptoms.append(row[feature].strip())
        
    diseases[disease].extend(symptoms)

diseases = {key: list(set(value)) for key, value in diseases.items()}
```
